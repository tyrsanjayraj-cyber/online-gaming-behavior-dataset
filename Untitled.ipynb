```python
import pandas as pd

df = pd.read_csv("../data/online_gaming_behavior_dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
retention = df.groupby("PlayerID")["SessionsPerWeek"].sum()

retention.head()
```

```python
import matplotlib.pyplot as plt

retention = df.groupby("PlayerID")["SessionsPerWeek"].sum()

plt.hist(retention)

plt.title("Player Retention Distribution")
plt.xlabel("Sessions per Player")
plt.ylabel("Number of Players")

plt.show()
```

```python
purchase = df.groupby("PlayerLevel")["InGamePurchases"].sum()

purchase.head()
```

```python
purchase.plot(kind="bar")

plt.title("In-Game Purchases by Player Level")
plt.xlabel("Player Level")
plt.ylabel("Total Purchases")

plt.show()
```

```python
leaderboard = df.groupby("PlayerID")["PlayTimeHours"].sum().sort_values(ascending=False)

top_players = leaderboard.head(10)

top_players
```

```python
top_players.plot(kind="bar")

plt.title("Top 10 Players by Play Time")
plt.xlabel("Player ID")
plt.ylabel("Play Time Hours")

plt.show()
```
